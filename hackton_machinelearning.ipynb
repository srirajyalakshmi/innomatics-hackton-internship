```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from category_encoders.cat_boost import CatBoostEncoder
from sklearn.naive_bayes import CategoricalNB
import pandas as pd

# Assuming you have a pandas DataFrame 'df' with categorical columns
# Make sure to replace this with your own dataset
# Example: df = pd.read_csv('your_dataset.csv')

# Specify your categorical columns and target column
categorical_columns = ['Review', 'Smell','portion']  # Replace with your actual column names
target_column = 'taste'  # Replace with your actual target column name

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(iris[categorical_columns], iris[target_column], test_size=0.2, random_state=42)

# Encode categorical features using CatBoostEncoder
encoder = CatBoostEncoder()
X_train_encoded = encoder.fit_transform(X_train, y_train)
X_test_encoded = encoder.transform(X_test)

# Create a Categorical Naive Bayes classifier
clf = CategoricalNB()

# Fit the classifier to the training data
clf.fit(X_train_encoded, y_train)

# Make predictions on the test set
y_pred = clf.predict(X_test_encoded)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy on the test set: {accuracy}")

# Display classification report
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
import requests
from bs4 import BeautifulSoup

def get_imdb_rating(movie_title):
    # IMDb search URL
    search_url = f"https://www.imdb.com/find?q={movie_title}"

    # Send a request to IMDb
    response = requests.get(search_url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find the first result (assuming it's the most relevant)
        result = soup.find('td', class_='result_text')

        if result:
            # Extract IMDb ID from the result
            imdb_id = result.find('a')['href'].split('/')[2]

            # IMDb movie URL
            movie_url = f"https://www.imdb.com/title/{imdb_id}/"

            # Send a request to the movie page
            movie_response = requests.get(movie_url)

            if movie_response.status_code == 200:
                movie_soup = BeautifulSoup(movie_response.text, 'html.parser')

                # Find the user rating element
                rating_element = movie_soup.find('span', itemprop='ratingValue')

                if rating_element:
                    # Extract the user rating
                    user_rating = rating_element.text
                    return user_rating
                else:
                    return "User rating not found on IMDb."
            else:
                return f"Failed to retrieve movie page for {movie_title}"
        else:
            return f"Movie not found on IMDb: {movie_title}"
    else:
        return f"Failed to perform IMDb search for {movie_title}"




# Scrape IMDb user ratings for each movie
for title in movies_list:
    rating = get_imdb_rating(title)
    print(f"{title}: {rating}")
```

```python
import requests
import numpy as np
from bs4 import BeautifulSoup

def scrapper(imdbId):
    id = str(int(imdbId))
    n_zeroes = 7 - len(id)
    new_id = "0"*n_zeroes + id
    URL = f"https://www.imdb.com/title/tt{new_id}/"
    request_header = {'Content-Type': 'text/html; charset=UTF-8', 
                      'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:109.0) Gecko/20100101 Firefox/119.0', 
                      'Accept-Encoding': 'gzip, deflate, br'}
    response = requests.BeautifulSoup(URL, headers=request_header)
    soup = BeautifulSoup(response.text)
    imdb_rating = soup.find('FILL_IN_THE_BLANK', attrs={'FILL_IN_THE_BLANK' : 'FILL_IN_THE_BLANK'})
    return imdb_rating.text if imdb_rating else np.nan
```

```python
import requests
import numpy as np
from bs4 import BeautifulSoup

def scrape_imdb_rating(imdb_id):
    imdb_url = f"https://www.imdb.com/title/tt{imdb_id}/"
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:119.0) Gecko/20100101 Firefox/119.0'}
    
    response = requests.get(imdb_url, headers=headers)

    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Find the element containing the IMDb rating
        rating_element = soup.find('span', itemprop='ratingValue')
        
        if rating_element:
            return rating_element.text.strip()
        else:
            return np.nan
    else:
        return np.nan

# Initialize variables to keep track of the highest rating and its corresponding IMDb ID
highest_rating = 0
highest_rated_movie_id = None

# Scrape IMDb ratings for each movie
for imdb_id in imdb_ids:
    rating = float(scrape_imdb_rating(imdb_id))
    
    if not np.isnan(rating) and rating > highest_rating:
        highest_rating = rating
        highest_rated_movie_id = imdb_id

# Print the movie with the highest rating
print(f"The movie with the highest IMDb rating is IMDb ID: {highest_rated_movie_id}, Rating: {highest_rating}")
```
